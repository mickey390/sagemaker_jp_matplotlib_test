## 日本語設定を行う

````python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

#日本語の基本設定 (スクリプト全体に影響する)
plt.rcParams['font.family'] = 'IPAGothic'
#or
# plt.rcParams['font.family'] = 'IPAMincho'

# フォントサイズの設定
plt.rcParams['font.size'] = 13

#  dpiの設定
plt.rcParams['figure.dpi'] = 100
````

## 日本語描画テスト(スクリプト全体パターン)

````python
plt.figure(1)
plt.plot([1,2,3],[5,2,1.1])
plt.xlabel('X軸')
plt.ylabel('Y軸')
````

## 日本語描画テスト(一部パターン)

````python
font_path = '/usr/share/fonts/ipa-mincho/ipam.ttf'
font_prop = fm.FontProperties(fname=font_path)
font_prop.set_style('normal')
font_prop.set_size('40')

plt.figure(2)
plt.plot([10,2,3],[9,5,6])
plt.xlabel('ふつうのX軸')
plt.ylabel('でかいY軸', fontproperties=font_prop)
````
